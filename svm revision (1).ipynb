```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs,make_moons
from sklearn.model_selection import train_test_split
```

```python
#linear svm for separable data
# svm svc for non separable data where u can specify kernel to convert to 3d
```

```python
X1,y1=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,class_sep=2.0,random_state=42)
```

```python
X1
```

```python
y1 #classes
```

```python
plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.show()
```

```python
X2,y2=make_blobs(n_samples=1000,centers=2,n_features=2,random_state=42)
```

```python
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.show()
```

```python
X3,y3=make_moons(n_samples=1000,noise=0.3,random_state=42) #noise is overlapping
```

```python
plt.scatter(X3[:,0],X3[:,1],c=y3)
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X1,y1,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import LinearSVC
clf=LinearSVC()
clf.fit(X_train,y_train)
```

```python
y_train
```

```python
clf.score(X_train,y_train) #to check training accuracy
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
